```python
import cv2
import matplotlib.pyplot as plt
import glob
import random
```

```python
def show_sample(img_file):
    img = cv2.imread(img_file)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    print(img_file)
    plt.imshow(img)
```

```python
print(f"Liczba zdjęć z zbiorze: {len(glob.glob('*'))}")

files=os.listdir("./")
random_img=random.choice(files)
show_sample(random_img)
```
